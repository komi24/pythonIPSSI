```python
import numpy as np
import matplotlib.pyplot as plt


morts = np.random.normal(10, 2, (100,2))
survivants = np.random.normal(40, 2, (100,2))

plt.scatter(morts[:,0], morts[:,1], c="r")
plt.scatter(survivants[:,0], survivants[:,1], c="g")
plt.show()

```

```python
from sklearn.svm import SVC


data_examples = np.concatenate([morts,survivants])
data_results = np.array([0] * 100 + [1] * 100)

data_examples.shape
data_results

classifier = SVC()
classifier.fit(data_examples, data_results)
```

```python
classifier.predict([[40, 40],[10, 10]])
```

```python
data_test = np.random.uniform(0,50,(500,2))
data_predict = classifier.predict(data_test)
# print(data_predict)
plt.scatter(data_test[:,0], data_test[:,1], c=data_predict, cmap="Set1")
plt.show()
```
