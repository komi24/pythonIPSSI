```python
# pourcentage des hommes qui ont survécu (par rapport au nombre d'hommes sur le bateau)
male_survivors_pct = len(df[(df["Sex"] == "male") & (df["Survived"] == 1)]) / len(df[df["Sex"] == "male"])
print(f"Pourcentage des hommes qui ont survécu {int(male_survivors_pct * 100)}%")
# pourcentage des femmes qui ont survécu (par rapport au nombre de femmes sur e batreau)
female_survivors_pct = len(df[(df["Sex"] == "female") & (df["Survived"] == 1)]) / len(df[df["Sex"] == "female"])
print(f"Pourcentage des femmes qui ont survécu {int(female_survivors_pct * 100)}%")
```
